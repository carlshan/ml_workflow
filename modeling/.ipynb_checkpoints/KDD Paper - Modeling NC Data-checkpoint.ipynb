```python
exclude_features = ['Grade_repeats', 
                    'nograd',
                    'RepeatedYears[T.2013, 2014]',
                     'RepeatedGrades[T.12]',
                     'RepeatedGrades[T.9, 11]',
                     'RepeatedYears[T.2013]',
                     'RepeatedYears[T.2010]',
                     'RepeatedGrades[T.10, 10]',
                     'RepeatedYears[T.2011, 2012]',
                     'RepeatedGrades[T.9, 9]',
                     'RepeatedYears[T.2012, 2013]',
                     'RepeatedGrades[T.10]',
                     'RepeatedGrades[T.9]',
                     'RepeatedYears[T.2011, 2013]',
                     'RepeatedYears[T.2011]',
                     'RepeatedGrades[T.10, 12]',
                     'RepeatedGrades[T.11]',
                     'RepeatedYears[T.2012]',
                     'RepeatedYears[T.2012, 2014]',
                     'RepeatedGrades[T.9, 12]'
                    ]
features = [f for f in set(train.columns).intersection(test.columns) if f not in exclude_features]
outcome = 'nograd'
```
