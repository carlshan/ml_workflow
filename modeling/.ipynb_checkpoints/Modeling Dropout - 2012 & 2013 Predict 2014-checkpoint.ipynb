```python
# Abstracted from model/classification.py 
def preprocess(dataSet, dependentVar, doFeatureSelection=True, doPCA=False, nComponents=10):
        """ Data pre-processing constructor.

        Constructor to pre-process pandas DataFrames, extracting and encoding the outcome
        labels (class), dropping them from the dataset and converting categorical variables
        into integer numbers for compatibility with scikit-learn.

        Parameters
        ----------
        dataSet : pd.DataFrame
            The entire dataset as loaded and parsed in the main program
        dependentVar : string
            A string denoting the column to be used as the class
        doFeatureSelection : bool
            A flag to denote whether or not to perform feature selection
        doPCA : bool
            A flag to denote whether or not to perform principle component analysis
        nComponents : int
            The desired number of principle components

        """
        # Encode nominal features to conform with sklearn
        for i,tp in enumerate(dataSet.dtypes):
            if tp == 'object':
                unique_vals, dataSet.ix[:,i]  = np.unique(dataSet.ix[:,i] , return_inverse=True)

        # Set the dependent variable (y) to the appropriate column
        y = dataSet.loc[:,dependentVar]

        # Transform that information to a format that scikit-learn understands
        # This may be redundant at times
        labels = preprocessing.LabelEncoder().fit_transform(y)

        # Remove the dependent variable from training sets
        X = dataSet.drop(dependentVar,1).values

        # Perform entropy-based feature selection
        if doFeatureSelection:
            print 'Performing Feature Selection:'
            print 'Shape of dataset before feature selection: ' + str(X.shape)
            clf = DecisionTreeClassifier(criterion='entropy')
            X = clf.fit(X, y).transform(X)
            print 'Shape of dataset after feature selection: ' + str(X.shape) + '\n'

        # Normalize values
        X = preprocessing.StandardScaler().fit(X).transform(X)

        # Collapse features using principal component analysis
        if doPCA:
            print 'Performing PCA'
            estimator = decomposition.PCA(n_components=nComponents)
            X = estimator.fit_transform(X)
            print 'Shape of dataset after PCA: ' + str(X.shape) + '\n'

        # Save processed dataset, labels and student ids
        
        return (X, labels)
```

```python
# Creating flag for whether a student completed 12th grade at Cabarrus
def return_completed_grade_12(row, indexed_data_files_with_grade, reporting_year_col_names):
    """
    This is a very hacky function that is meant to be applied to a data frame, and should
    honestly only be run in the initial data transformation stage.

    Returns
    ----------
    0 or 1

    Parameters
    ----------
    indexed_data_files_with_grade : list
        List of csvs to search through. Should have a grade and reporting year column
    reporting_year_col_names : list
        List of names for reporting year in the csvs
    """
    student_id = row['External_Student_ID']
    for data_file in indexed_data_files_with_grade:
        grade_col = 'grade' if 'grade' in data_file.columns else 'Grade'
        reporting_year_col = [name for name in reporting_year_col_names if name in data_file.columns][0]
        filtered_rows = data_file.ix[[student_id]]
        if 12 in filtered_rows[grade_col].values:
            return 1
    return 0
```

```python
def get_retained(row, indexed_data_files_with_grade, reporting_year_col_names):
    """
    This is a very hacky function that is meant to be applied to a data frame, and should
    honestly only be run in the initial data transformation stage.
    It returns tuple where the first element is a 0,1 flag if a student was retained, and the second element is a list of the grades they were retained in

    Returns
    ----------
    grad_on_time : [0, 1]

    Parameters
    ----------
    indexed_data_files_with_grade : list
        List of csvs to search through. Should have a grade and reporting year column
    reporting_year_col_names : list
        List of names for reporting year in the csvs
    """
    student_id = row['External_Student_ID']
    expected_grad_count = {'yes': 0,
                           'no': 0,}
    expected_graduating_year = row['expected_graduation_year_majority_vote']
    
    for data_file in indexed_data_files_with_grade:
        grade_col = 'grade' if 'grade' in data_file.columns else 'Grade'
        reporting_year_col = [name for name in reporting_year_col_names if name in data_file.columns][0]
        filtered_rows = data_file.ix[[student_id]]
        grades_in_expected_graduating_year = filtered_rows[filtered_rows[reporting_year_col] == expected_graduating_year][grade_col].values
        if 12 in grades_in_expected_graduating_year:
            expected_grad_count['yes'] += 1
        else:
            expected_grad_count['no'] += 1
    if expected_grad_count['yes'] >= retained_count['no']:
        grad_on_time = 1
    else:
        grad_on_time = 0
    return grad_on_time
```
