```python
def make_rename_dict(cols, graduating_year):
    """Maps each year to a grade, depending on graduating year"""
    d = {col: col for col in cols}
    mappings = zip(range(8, 13), range(graduating_year-4, graduating_year+1))
    for grade, year in mappings:
        append = str(grade) + 'th_grade'
        temp = {col: col.replace(str(year), append) for col in cols if str(year) in col}
        d.update(temp)
    return d
```

```python
d = make_rename_dict(cohort_a_merged, 2012)
```

```python
d
```

```python
renamed_cohort_a = cohort_a_merged.rename(columns=make_rename_dict(cohort_a_merged, 2012))
renamed_cohort_b = cohort_b_merged.rename(columns=make_rename_dict(cohort_b_merged, 2013))
renamed_cohort_c = cohort_c_merged.rename(columns=make_rename_dict(cohort_c_merged, 2014))
```

```python
dropouts = dropout_data[~dropout_data['External_Student_ID'].isin(exclude)]['External_Student_ID']
```

```python
# Making the dropout flag
make_dropout = lambda row: 1 if row['External_Student_ID'] in dropouts else 0
renamed_cohort_a['dropout'] = renamed_cohort_a.apply(make_dropout, axis=1)
renamed_cohort_b['dropout'] = renamed_cohort_b.apply(make_dropout, axis=1)
renamed_cohort_c['dropout'] = renamed_cohort_c.apply(make_dropout, axis=1)
```
