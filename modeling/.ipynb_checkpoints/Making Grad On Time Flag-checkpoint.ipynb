```python
raw_data_path = "s3://dsapp-education/NC-Cabarrus/raw_data/"
clean_data_path = 's3://dsapp-edu-data/NC-Cabarrus/cleaned_data/'
```

```python
dfAddHist = pd.read_table(raw_data_path + 'Student_Address_History.txt')
dfDemogAttend = pd.read_table(raw_data_path + 'Student_Demographic_Attendance.txt')
dfGPADet = pd.read_table(raw_data_path + 'Student_GPA_Detail.txt')
dfGPAYearSum = pd.read_table(raw_data_path + 'Student_GPA_YearlySummary.txt')
dfSuspensionData = pd.read_table(raw_data_path + 'Student_Suspension_Data_20141001.txt')
```

```python
data = pd.read_csv(clean_data_path + 'master_10_14_2014.csv')
```

```python
indexed_addhist = dfAddHist.set_index('External_Student_ID')
indexed_demogattend = dfDemogAttend.set_index('External_Student_ID')
indexed_gpayearly = dfGPAYearSum.set_index('External_Student_ID')
indexed_gpadetailed = dfGPADet.set_index('External_Student_ID')
indexed_susp = dfSuspensionData.set_index('External_Student_ID')
```

```python
indexed_susp['Converted Reporting Year'] = indexed_susp['Incident Date'].apply(util.convert_to_reporting_year)
```

```python
reporting_year_col_names = ['ReportingYear', 'reporting_year', 'RportingYear', 'Converted Reporting Year']
```

```python
indexed_data_files_with_grade = [indexed_addhist, 
                                 indexed_demogattend, 
                                 indexed_gpadetailed,
                                 indexed_susp,]
```

```python
def get_grad_on_time_majority_vote(row, indexed_data_files_with_grade, reporting_year_col_names):
    """
    This is a very hacky function that is meant to be applied to a data frame, and should
    honestly only be run in the initial data transformation stage.
    It returns tuple where the first element is a 0,1 flag if a student graduated on time

    Returns
    ----------
    grad_on_time : [0, 1]

    Parameters
    ----------
    indexed_data_files_with_grade : list
        List of csvs to search through. Should have a grade and reporting year column
    reporting_year_col_names : list
        List of names for reporting year in the csvs
    """
    student_id = row['External_Student_ID']
    expected_grad_count = {'yes': 0,
                           'no': 0,}
    expected_graduating_year = row['expected_graduation_year_majority_vote']
    
    for data_file in indexed_data_files_with_grade:
        grade_col = 'grade' if 'grade' in data_file.columns else 'Grade'
        reporting_year_col = [name for name in reporting_year_col_names if name in data_file.columns][0]
        filtered_rows = data_file.ix[[student_id]]
        grades_in_expected_graduating_year = filtered_rows[filtered_rows[reporting_year_col] == expected_graduating_year][grade_col].values
        if 12 in grades_in_expected_graduating_year:
            expected_grad_count['yes'] += 1
        else: # nan, or other values
            expected_grad_count['no'] += 1
    if expected_grad_count['yes'] >= expected_grad_count['no']:
        grad_on_time = 1
    else:
        grad_on_time = 0
    return grad_on_time
```

```python
data['graduated_on_time_majority_vote'] = data.apply(get_grad_on_time_majority_vote,
                                                     indexed_data_files_with_grade=indexed_data_files_with_grade,
                                                     reporting_year_col_names=reporting_year_col_names,
                                                     axis=1)
```

```python
data.to_csv('master.csv', index=False)
```

```python
# Making transfer out flag
# Transfer out if you did not complete 12th grade in cabarrus but you are also not a dropout
def create_transfer_out(row):
    completed_in_cabarrus = row['completed_grade_12_in_cabarrus']
    dropout_outcome = row['dropout_outcome']
    if (dropout_outcome + completed_in_cabarrus) == 0:
        return 1
    else:
        return 0
```

```python
data['transferred_out_before_graduating'] = data.apply(create_transfer_out,
                                                       axis=1)
```

```python
def create_fail_to_finish_within_4(row):
    dropout_outcome = row['dropout_outcome']
    grad_on_time = row['graduated_on_time_majority_vote']
    transfer_status = row['transferred_out_before_graduating']        
    if (dropout_outcome == 1) or (grad_on_time == 0):
        if transfer_status == 1: # we don't know what happened to transfers
            return nan
        else:
            return 1
    else:
        return 0
```

```python
data['fail_to_finish_high_school_in_4_years'] = data.apply(create_fail_to_finish_within_4, axis=1)
```
