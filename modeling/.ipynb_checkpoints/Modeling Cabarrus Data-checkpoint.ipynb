## Modeling 6th Grade Data

```python
train = pd.read_csv(datapath + 'train_08.csv')
```

```python
test = pd.read_csv(datapath + 'test_08.csv')
```

```python
# Changing min_samples_leaf affects the marginal precision curve A LOT
# Leaving it at 1 gets a nice-looking marginal precision curve
# This makes sense, since if we have min_samples_leaf being set at 25 or so
# then, there are many possible probabilities that can come as the average
# of many trees. E.g., Averaging over 3 trees predicting that individually 
# predicted probabilities (16/25, 22/25, 14/25) can generate many unique numbers
# So in plotting the marginal precision curve, only a very small number of
# points will have that exactly probability. 
# On the other hand, if min_samples_leaf = 1, then the predicted probabilities
# of the average of 3 threes is likely to be 0/3, 1/3, 2/3 or 3/3.
clf = RandomForestClassifier(n_estimators=100, criterion='entropy',
                             min_samples_leaf=1)
```

```python
features = [f for f in set(train.columns).intersection(test.columns) if f != 'nograd']
outcome = 'nograd'
```

```python
fitted_model, predicted_probs = workflow_model_setup.run_model(train,
                                                               test,
                                                               features,
                                                               outcome,
                                                               clf,
                                                               'RF-100-gini',)
```

```python
predicted_probs = [p[1] for p in predicted_probs]
```

```python
hist(predicted_probs)
```

```python
pr_auc, precision, recall, thresholds = workflow_diagnostics.plot_precision_recall_curve(test[outcome], 
                                                 predicted_probs,
                                                 'RF-100-gini')
```

```python
roc_auc, fpr, tpr, thresholds = workflow_diagnostics.plot_roc_curve(test[outcome],
                                                                    predicted_probs,
                                                                    'RF-100-gini')
```

```python
importances, std = workflow_diagnostics.plot_feature_importance(fitted_model, features)
```

```python
marginals = workflow_diagnostics.plot_marginal_precision_curve(test[outcome], 
                                                               predicted_probs,
                                                               marker='o',
                                                               ls='None')
```
